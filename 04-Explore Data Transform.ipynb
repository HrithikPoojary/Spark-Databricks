```python
spark.sql("""select weekofyear(to_timestamp(calldate , 'MM/dd/yyyy')) as week_year , count(*) as count from dev.spark_db.sf_fire_calls
                where year(to_timestamp(calldate , 'MM/dd/yyyy')) == 2018
                group by week_year
                order by count desc
          """).display()
```

```python
from pyspark.sql.functions import weekofyear,year , to_timestamp,col

df.where(year(to_timestamp(df.CallDate ,'MM/dd/yyyy')) == 2018)\
  .select(weekofyear(to_timestamp(df.CallDate ,'MM/dd/yyyy')).alias('week_year'))\
  .groupBy(col('week_year')).count().alias('count')\
  .orderBy(col('count').desc())\
  .show()
```

```python
df.where("year(to_timestamp(CallDate ,'MM/dd/yyyy')) = 2018")\
  .selectExpr("weekofyear(to_timestamp(CallDate ,'MM/dd/yyyy')) as week_year")\
  .groupBy(col('week_year')).count().alias('count')\
  .orderBy(col('count').desc())\
  .show()
```
