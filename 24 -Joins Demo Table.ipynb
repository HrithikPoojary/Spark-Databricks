```python
from pyspark.sql.types import StructType,StructField , IntegerType , StringType ,DoubleType,DateType , TimestampType


facilities_schema = StructType(
    [

    StructField('facility_id' ,  IntegerType()),
    StructField( 'facility_name', StringType()), 
    StructField( 'member_cost', DoubleType() ),
    StructField( 'guest_cost', DoubleType()), 
    StructField('initial_outlay', DoubleType()), 
    StructField('monthly_maintainance', DoubleType())

    ]

)
```

```python
members_schema = StructType(
    [

    StructField('member_id' ,  IntegerType()),
    StructField( 'last_name', StringType()), 
    StructField( 'first_name', StringType() ),
    StructField( 'address', StringType()), 
    StructField('zip_code', StringType()), 
    StructField('telephone', StringType()),
    StructField('recommended_by', StringType()),
    StructField('joining_date', DateType())
    ]

)
```

```python
bookings_schema = StructType(
    [

    StructField('booking_id' ,  IntegerType()),
    StructField( 'facility_id', IntegerType()), 
    StructField( 'member_id', IntegerType() ),
    StructField( 'start_time', TimestampType()), 
    StructField('slots', IntegerType())
    ]
)
```

```python
facilities_df = spark.read.format("csv")\
                .option("header" , "true")\
                    .schema(facilities_schema)\
                        .load(path = '/Volumes/dev/spark_db/datasets/spark_programming/data/facilities.csv')

members_df = spark.read.format("csv")\
                    .option("header" , "true")\
                        .schema(members_schema)\
                            .load(path = '/Volumes/dev/spark_db/datasets/spark_programming/data/members.csv')
bookings_df = spark.read.format("csv")\
                        .option("header" , "true")\
                            .schema(bookings_schema)\
                                .load(path = '/Volumes/dev/spark_db/datasets/spark_programming/data/bookings.csv')

facilities_df.write.mode("overwrite").option("overwriteSchema" , "true").saveAsTable("dev.spark_db.facilities")
members_df.write.mode("overwrite").saveAsTable("dev.spark_db.members")
bookings_df.write.mode("overwrite").saveAsTable("dev.spark_db.bookings")
```
