```python
data_lst = [
    (101 , 'Mumbai' , 'Goa' , 587),
    (102 , 'Mumbai' , 'Bangalore' , 985),
    (102 , 'Mumbai' , 'Bangalore' , 985),
    (103 , 'Dehli' , 'Chennai' , 2208),
    (104 , 'Dehli' , 'Chennai' , 2208),
    (105 , 'Bangalore' , 'Kolkata' , 1868),
    (105 , 'Bangalore' , 'Kolkata' , 185)
]
schema_lst = 'id int , source string , destination string ,distance int'

df = spark.createDataFrame(data = data_lst , schema = schema_lst)
df.show()
```

```python
from pyspark.sql.functions import col

df.where("source='Mumbai' and destination = 'Bangalore'").show()
df.where( (col('source') == 'Mumbai') & (col('destination') == 'Bangalore')).show()
df.where( (df.source == 'Mumbai') & (df.destination == 'Bangalore')).show()
```

```python
df1 = df.distinct()
df.exceptAll(df1).show()
```

```python
df.distinct().show()
```

```python
df.dropDuplicates(["id" , "source" , "destination"]).show()
```
