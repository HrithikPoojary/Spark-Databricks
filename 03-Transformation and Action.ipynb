```python
from pyspark.sql import SparkSession

spark = SparkSession.builder\
        .appName('Transformation and Actions')\
        .getOrCreate()
```

```python
spark.sql("""select calltype,zipcode ,count(*) from dev.spark_db.sf_fire_calls
            where calltype is not null
             group by calltype,Zipcode
             order by count(*) desc 
             limit 3
          """).display()
```

```python
df = (
    spark.read.format("csv").
    option('header', 'true').
    option('inferSchema','true').
    load('/Volumes/dev/spark_db/datasets/spark_programming/data/sf-fire-calls.csv')
)
```

```python
df.display()
```

```python
from pyspark.sql.functions import col

df.where("CallType is not Null")\
    .groupBy('CallType', 'Zipcode').count().alias('count')\
    .orderBy(col('count').desc())\
        .limit(3).display()
```

```python
# Using existing table

df2 = spark.read.table("dev.spark_db.sf_fire_calls")
```

```python
from pyspark.sql.functions import col
df2.where("calltype is not null").groupBy(df2.CallType , df2.Zipcode).count().alias("count").orderBy(col('count').desc()).limit(3).display()
```

```python
df_1 = df.select(df.CallType,df.Zipcode)
df_2 = df_1.where(df_1.CallType.isNotNull())
df_3 = df_2.groupBy(df_2.CallType,df_2.Zipcode).count().alias('count')
df_4 = df_3.orderBy(col('count'),ascending = False)
df_5 = df_4.limit(3)  
```
