```python
df = spark.read.format("csv")\
               .option("header" , "true")\
               .option("inferSchema" , "true")\
               .load(path = '/Volumes/dev/spark_db/datasets/spark_programming/data/invoices.csv')
df.show(5)
```

```python
from pyspark.sql.functions import round,expr
df.withColumn("total_price" , round(df.Quantity * df.UnitPrice ,2 )).show(5)
df.withColumn("total_price" , expr(" round(Quantity * UnitPrice , 2)")). show(5)
```

Save to variable Variable 

```python
from pyspark.sql.functions import col
variable_1 = round(col("quantity") * col("unitprice") , 2)
df.withColumn("total_price" , variable_1).show(5)
```

```python
variable_2 = expr("round(quantity * unitprice , 2)")
df.withColumn("total_price" , variable_2).show(5)
```

```python
df.describe(['invoiceno' ,'quantity' , 'unitprice']).show()
```

```python
df.summary().select('summary' ,'invoiceno' ,'quantity' , 'unitprice').display()
```

```python
from pyspark.sql.functions import min,max,percentile

df.select(min('unitprice') , max('unitprice') , percentile('unitprice' , 0.99)).show()
```
