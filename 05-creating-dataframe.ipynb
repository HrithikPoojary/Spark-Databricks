```python
# Read using connector

df = (

    spark.read.format('csv')
    .option("header" , 'true')
    .option("inferSchema" , 'true')
    .load(path = '/Volumes/dev/spark_db/datasets/spark_programming/data/sf-fire-calls.csv')
)

df.printSchema()
```

```python
spark.read.table("dev.spark_db.sf_fire_calls").printSchema()
```

```python
df = spark.sql("""select callnumber,zipcode,calldate from dev.spark_db.sf_fire_calls
                  limit 5
               """)
df.show()
df.printSchema()
```

```python
table_name   = 'dev.spark_db.sf_fire_calls'
df = spark.sql(f"""select callnumber,zipcode,calldate from {table_name}
                  limit 5
               """)
df.show()
df.printSchema()
```

```python
lst = [('Raju' , 20) , ('Jaggu' , 15)]
schema = 'name string , age int'

df = spark.createDataFrame(data = lst , schema = schema )
df.show()
df.printSchema()
```
