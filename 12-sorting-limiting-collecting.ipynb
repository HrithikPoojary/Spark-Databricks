```python
flight_time_df = spark.read.table("dev.spark_db.flight_time") 
```

```python
flight_time_df.limit(3).display()
```

```python
from pyspark.sql.functions import col

flight_time_df.where(" fl_date = '2000-01-16' and origin = 'AUS' and dest = 'ORD'")\
    .selectExpr("arr_time - crs_arr_time as delayed")\
    .orderBy(col("delayed").desc())\
    .limit(3)\
    .show(truncate = False)
```

```python
from pyspark.sql.functions import col,expr

flight_time_df.where(" fl_date = '2000-01-16' and origin = 'AUS' and dest = 'ORD'")\
    .withColumn("delayed" , expr("arr_time - crs_arr_time"))\
    .orderBy(col("delayed").desc())\
    .limit(3)\
    .display()
```

```python
# Pulling dataframe record to python varial is called collecting record(pull the data to client)
# collecting records from the distributed computer , it will cost you memory issue in client machine due to large value of the
# stored in the variable.

from pyspark.sql.functions import col,expr

flight_time_df_1 = flight_time_df.where(" fl_date = '2000-01-16' and origin = 'AUS' and dest = 'ORD'")\
    .withColumn("delayed" , expr("arr_time - crs_arr_time"))\
    .orderBy(col("delayed").desc())\
    .limit(3)

python_lst = flight_time_df_1.collect()
```

```python
for i in python_lst:
    print(i)
```

```python
for i in python_lst:
    print(f"{i.OP_CARRIER} is {i.delayed} minutes late")
```

```python
# convert row object to dict of list
#https://spark.apache.org/docs/latest/api/python/reference/pyspark.sql/row.html

top_3_list = [row.asDict()  for row in python_lst]

top_3_list
```

```python
for i in top_3_list:
    print(f"{i['OP_CARRIER']} flight is {i['delayed'].total_seconds()/60} minutes late")
```

```python
# to 3rd result

from pyspark.sql.functions import col,expr

flight_time_df.where(" fl_date = '2000-01-16' and origin = 'AUS' and dest = 'ORD'")\
    .withColumn("delayed" , expr("arr_time - crs_arr_time"))\
    .orderBy(col("delayed").desc())\
    .limit(3)\
    .offset(2)\
    .select(col("delayed"))\
    .display()
```
