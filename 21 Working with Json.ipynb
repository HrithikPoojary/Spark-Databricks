```python
from pyspark.sql.types import StructField , StructType , StringType , ArrayType , MapType

schema = StructType(
    [
        StructField(name = 'ID' , dataType = StringType() ,nullable = True),
        StructField(name = 'FirstName' , dataType = StringType() ,nullable = True),
        StructField(name = 'LastName' , dataType =StringType() ,nullable = True),
        StructField(name = 'Address' , dataType = StructType( [
                                                    StructField('AddressLine1' , StringType() , nullable = True ),
                                                    StructField('AddressLine2' , StringType() ,nullable = True ),
                                                    StructField('City' , StringType() ,nullable = True ),
                                                    StructField('State' , StringType() ,nullable = True ),
                                                    StructField('Country' , StringType() ,nullable = True ),
                                                    StructField('Pin' ,  StringType() ,nullable = True) 
                                                    ] 
                                                    )
                                , nullable = True
                    ),
        StructField(name = 'Skills' ,dataType =  ArrayType(elementType = StructType(
                                                                    [
                                                                    StructField('Skill'  , StringType() ,nullable = True),
                                                                    StructField('YearsOfExperience' , StringType() ,nullable = True)
                                                                    ]   
                                                                    )
                                                    ,containsNull = True
                                        )
                              ,nullable = True
                    ),
        StructField(name = 'Contacts', dataType =  MapType(keyType = StringType() , valueType = StringType() , valueContainsNull = True) , nullable = True)
    ]
)
```

```python
df = spark.read.format('json')\
                .schema(schema)\
                    .load(path = '/Volumes/dev/spark_db/datasets/spark_programming/data/students_online.json')
df.write.mode('overwrite').saveAsTable('dev.spark_db.online_students')
```

```python
spark.sql(

    '''
    select address['country'] , count(id)  as count from dev.spark_db.online_students
    group by address['country']
    '''
).display()
```

```python
spark.sql('''
          

          select id , skills.* from (
          select id , explode(Skills) as Skills from dev.spark_db.online_students
          )
          where Skills['Skill'] like '%Spark%' and Skills['YearsOfExperience'] > 2 
          ''').display()
```
