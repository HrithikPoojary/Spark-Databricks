```python
from pyspark.sql import SparkSession

spark = SparkSession.builder\
        .appName('Spark')\
            .getOrCreate()
```

```python
data_lst = [ (2022 , 5, 18) , (9999,12,31) ,(-9999,1,1),(10000 ,1,1) ,
            (-10000,1,1) , (193 ,5 ,25) ,(99,5,25) , (1000,2 ,29)   ]
schema = ['Y' , 'M' , 'D']

df = spark.createDataFrame(data = data_lst , schema = schema )

from pyspark.sql.functions import concat_ws,col
df = df.withColumn('date_str' , concat_ws('-' , col('Y') , col('M') , col('D')))
df.show()
```

```python
from pyspark.sql.functions import expr

df = df.withColumn('valid_date' , expr( "try_to_date(date_str , 'y-M-d')")).drop('Y','M','D')
# 'y-M-d' = y - any number year like 1,193,99 etc M = any number of month , d =  any number of date

df.show()
```

```python
from pyspark.sql.functions import date_add , date_sub , add_months

df.withColumns(
    {
         'add_5_days' : date_add(col('valid_date') , 5) ,
         'sub_5_days' : date_sub(col('valid_date') , 5) ,
         'add_5_months' : add_months(col('valid_date') , 5),
         'sub_5_months' : add_months(col('valid_date') , -5)
      }
).show()
```
