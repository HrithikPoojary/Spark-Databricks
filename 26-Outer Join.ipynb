```python
bookings_df = spark.table("dev.spark_db.bookings")
```

```python
members_df = spark.table("dev.spark_db.members")
```

```python
facilities_df = spark.table("dev.spark_db.facilities")
```

```python
spark.sql("""
          
          select m.member_id , first_name , last_name , address , facility_id , slots  from dev.spark_db.members m
          left join dev.spark_db.bookings b
          on m.member_id = b.member_id
          where first_name = 'Darren' and last_name = 'Smith'
          order by slots desc nulls first
          
          """).display()
```

```python
from pyspark.sql.functions import col

members_df.join(bookings_df , on = members_df.member_id == bookings_df.member_id , how = 'left' )\
            .where( (members_df.first_name == 'Darren') & (members_df.last_name == "Smith")  )\
                .select(members_df.member_id, "first_name" , "last_name" , "address" , "facility_id" , "slots")\
                    .orderBy(col("slots").desc_nulls_first())\
                        .display()
```

```python
bookings_df = spark.table("dev.spark_db.bookings").alias("b")
members_df = spark.table("dev.spark_db.members").alias("m")
facilities_df = spark.table("dev.spark_db.facilities").alias("f")
```

```python
from pyspark.sql.functions import expr,col

members_df.join(bookings_df , on = expr("m.member_id = b.member_id") , how = 'left')\
            .join(facilities_df , on = expr("f.facility_id = b.facility_id") , how = 'left')\
                .where("m.first_name = 'Darren' and m.last_name = 'Smith'")\
                    .select("facility_name" , 
                            "slots" ,
                            expr("member_cost * slots as booking_amount") ,
                            "start_time" , 
                            members_df.member_id  , 
                            expr("first_name ||' ' || last_name as member_name") ,
                            "telephone" ,
                            "address")\
                    .orderBy(col("slots").desc_nulls_first())\
                    .display()
```
