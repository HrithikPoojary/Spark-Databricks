```python
data_lst = [
    (100,"Prasant" , 30),
    (101 , "David" , None),
    (102 , "Sushant" , None),
    (103 , "Abdul" , 45),
    (104 , "Shruti" , 28)
]

schema_lst = 'id int , name string , age int'

df = spark.createDataFrame(data = data_lst , schema = schema_lst)

df.display()
```

```python
from pyspark.sql.functions import col
df.where(col('age') == 28).show()
df.where("age = 28").show()
```

```python
from pyspark.sql.functions import isnull
df.where(df.age.isNull()).show()
df.where(" age is null").show()
df.where(isnull(df.age)).show()
```

```python
from pyspark.sql.functions import expr
df.withColumn("age>null" , expr("age > 29")).display()
df.withColumn("age=null" , expr("age = 29")).display()
```

```python
df.where(df.age > 29).show()
```

```python
from pyspark.sql.functions import coalesce , lit

df.withColumn("decreaseAge" , expr("nvl( age , 23 ) -23")).show()
df.withColumn("decreaseAge" , coalesce(df.age , lit(23)) - 23 ).show()
```

```python
df.selectExpr("avg(age) as avg").show()
```
