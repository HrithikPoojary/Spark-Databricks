```python
from pyspark.sql.functions import col,expr

bookings_df = spark.table("dev.spark_db.bookings").where("slots>5")
members_df = spark.table("dev.spark_db.members").where("last_name == 'Smith'")
facilities_df = spark.table("dev.spark_db.facilities")

# for multiple join keys["memeber_id" , facility_id]
bmf = bookings_df.join(members_df , "member_id" , "inner")\
                .join(facilities_df , "facility_id" , "inner")\
                .selectExpr("member_id" , "first_name", "last_name" ,"facility_name" , "slots" , "slots*member_cost as booking_amount" ,"start_time")\
                .orderBy(col("first_name").asc() , col("booking_amount").desc())\
                .display()
```

```python
from pyspark.sql.functions import col,expr

bookings_df = spark.table("dev.spark_db.bookings").where("slots>5")
members_df = spark.table("dev.spark_db.members").where("last_name == 'Smith'")
facilities_df = spark.table("dev.spark_db.facilities")

# for multiple join keys["memeber_id" , facility_id]
bmf = bookings_df.join(members_df , "member_id" , "inner")\
                .join(facilities_df , "facility_id" , "inner")\
                .orderBy(col("first_name").asc() , expr("slots * member_cost").desc())\
                .selectExpr("member_id" , "first_name", "last_name" ,"facility_name" , "slots" , "slots*member_cost as booking_amount" ,"start_time")\
                .display()
```

```python
members_df.alias("m").join(members_df.alias("r") , expr("m.member_id == r.recommended_by") , 'inner')\
    .select("m.first_name" ,"m.last_name" , expr("r.first_name || r.last_name as recommended_by" ))\
        .display() 
```

```python
from pyspark.sql.functions import concat_ws

members_df.alias("m").join(members_df.alias("r") , expr("m.member_id == r.recommended_by") , 'inner')\
    .select("m.member_id" ,
            concat_ws(' ' , col("m.first_name") , col("m.last_name")).alias("member_name"), 
         concat_ws(' ' , col("r.first_name") , col("r.last_name")).alias("recommended_by"))\
        .display()
```
