```python
df = spark.read.table("dev.spark_db.sf_fire_calls")
```

```python
from pyspark.sql.functions import col
df = df.select(df.CallType , df.Zipcode)\
        .groupBy(col('CallType') , col('Zipcode')).count().alias('count')\
            .orderBy(col('count').desc())\
                .limit(5)
```
