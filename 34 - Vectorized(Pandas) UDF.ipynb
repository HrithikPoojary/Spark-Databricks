```python
from pyspark.sql.functions import pandas_udf
import pandas as pd

@pandas_udf(returnType = "string")
def member_type_pudf(id :pd.Series )-> pd.Series:
    return id.apply(lambda x: "Guest" if x == 0 else "Member")
```

```python
df = spark.table("dev.spark_db.members")

df.where(member_type_pudf("member_id") == 'Guest').display()
```

```python
df = spark.table("dev.spark_db.bookings")

df.where(member_type_pudf("member_id") == 'Guest').display()
```

```python
spark.udf.register("type_member" , member_type_pudf)
```

```python
spark.sql("""
          
          select * from dev.spark_db.bookings
          where type_member(member_id) == 'Guest'
          
          """).display()
```

```python
from pyspark.sql.functions import expr
df.where(expr("type_member(member_id) == 'Guest' ")).display()
```
