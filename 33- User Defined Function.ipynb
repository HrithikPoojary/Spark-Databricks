```python
from pyspark.sql.functions import udf


# Pickle defaule , arrow is bit fast
@udf(returnType = "string" , useArrow = True)
def member_type_udf(id :str):
    return "Guest" if id == 0 else "Member"
```

```python
df = spark.table("dev.spark_db.members")

df.withColumn("member_type" , member_type_udf("member_id")).display()
```

```python
spark.sql('''
          
          select * , member_type_udf(member_id) from dev.spark_db.members
          
          ''')
```

```python
from pyspark.sql.functions import expr

df.withColumn("member_type" , expr("member_type_udf(member_id)")).display()
```

```python
spark.udf.register("member_type_udf_sql" , member_type_udf)
```

```python
spark.sql("""
          
          select *, member_type_udf_sql(member_id) as member_type from dev.spark_db.members 
          
          """).display()
```

```python
from pyspark.sql.functions import expr


df.withColumn("member_type" , expr("member_type_udf_sql(member_id)")).display()
```
